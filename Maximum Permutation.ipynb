# Maximum Permutation <br>
https://www.hackerrank.com/contests/university-codesprint-4/challenges/maximum-permutation/copy-from/1306121061 <br>
Given an array search it or any of its permutations in another, bigger, array. Output the permutation with the most occurences in the bigger string. <br>
Basically some Karp Rabin thingy, I created a rolling hash ( a hash that has operations for appending one element and removing 1st element ), where the order of the characters is not important. I count the occurences of each permutation with a dict.


```
import os
import sys

def get_hash(string):
    hash_w = [0]*30
    hsh = 0
    for el in string:
        #if el not in hash_w:
        #    hash_w[ord(el)] = 0
        norm_ascii = ord(el) - 97
        hash_w[norm_ascii] += 1
        hsh += norm_ascii
    return hash_w, hsh

# Complete the function below.
def maximumPermutation(w, s):
    # Return the string representing the answer.
    hash_w_list, hash_w = get_hash(w)
    m = len(w)
    hash_permut_list, hash_permut = get_hash(s[:m])
    permutation_count = {}
        
    if hash_permut == hash_w:
        if hash_permut_list == hash_w_list:
            if s[:m] not in permutation_count:
                permutation_count[s[:m]] = 0
            permutation_count[s[:m]] += 1
        
    for i in range(len(s)-m):
        #roll the hash
        
        #old char
        norm_ascii = ord(s[i])-97
        hash_permut_list[norm_ascii] -= 1
        hash_permut -= norm_ascii
        
        #new char
        norm_ascii = ord(s[i+m])-97
        hash_permut_list[norm_ascii] += 1
        hash_permut += norm_ascii
        
        #check whether hashes match
        if hash_permut == hash_w:
            if hash_permut_list == hash_w_list:
                perm = s[i+1:i+m+1]
                if perm not in permutation_count:
                    permutation_count[perm] = 0
                permutation_count[perm] += 1
                
    max_count = 0
    max_str = ''
    for perm in permutation_count:
        count = permutation_count[perm]
        if count > max_count:
            max_count = count
            max_str = perm
        elif count == max_count:
            if perm < max_str:
                max_str = perm
    if max_count == 0:
        max_str = "-1"
    return max_str
```

```
#Hackerrank main
if __name__ == '__main__':
    f = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_i in range(t):
        w = input()

        s = input()

        result = maximumPermutation(w, s)

        f.write(result + "\n")

    f.close()
```
