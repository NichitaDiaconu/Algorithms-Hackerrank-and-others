# Solution:
We first compute the for each index the sum up to that index modulo m
store the sums in sums
For each element in sums the candidates of the best sum are:
* i) the sum from 0 up to that index: sums[i]
* ii) the sum from some index j up to i; this sum is only relevant if sums[j] is the smallest, higher value than sums[i]

Computing ii) is the tricky part which has to be reduced to log(n). Searching for the element with binary search and inserting new elements with list.insert() had the best result.. which I expect to be O(n) for insertion and to do worse than AVL insert + search, but...

```
def binary_search_interface(array, value):
    return binary_search_rightmost_index(array, value, 0, len(array))

def binary_search_rightmost_index(array, value, start, stop):
    while start < stop:
        mid = int((start + stop) / 2)
        if value < array[mid]:
            stop = mid
        else:
            start = mid + 1
    return start
  
def maximum_sum(a, m):
    #The better implementation with ordered list, not AVL as DS...
    #actually much better time
    #so much to learn
    
    #We first compute the for each index the sum up to that index modulo m
    #store the sums in sums
    sums = [0] * len(a)
    for idx in range(len(a)):
        sums[idx] = (sums[idx - 1] + a[idx]) % m
    
    result = 0
    ordered_list = []
    for i in range(len(sums)):
         #For each element in sums the candidates of the best sum are:
         #   i) the sum from 0 up to that index: sums[i]
         #   ii) the sum from some index j up to i; this sum is only relevant if sums[j] is the smallest, higher value than sums[i]
        j = binary_search_interface(ordered_list, sums[i])
        
        if sums[i] > result:
            result = sums[i]
        
        if j < i:
            #if j is not the index of the highest element in the list
            candidate = (sums[i] - ordered_list[j]) % m
            if candidate >  result:
                result = candidate
        ordered_list.insert(j, sums[i])
        
    return result
    
if __name__ == "__main__":
    q = int(input().strip())
    for a0 in range(q):
        n, m = input().strip().split(' ')
        n, m = [int(n), int(m)]
        a = list(map(int, input().strip().split(' ')))
        result = maximum_sum(a, m)
        print(result)
```
