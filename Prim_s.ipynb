```
#!/bin/python3

import sys

def insert_queue(queue, pair_insert):
    #could be binary search...
    for idx, pair in enumerate(queue):
        if pair_insert[1] < pair[1]:
            queue.insert(idx, pair_insert)
            return
    queue.append(pair_insert)
    return

def prims(n, graph, start):
    # Complete this function
    queue = [(start,0)]
    min_sum = 0
    while len(graph) != 0:
        min_neigh, min_weight = queue[0]
        
        if min_neigh not in graph:
            #vertex was already visited
            queue = queue[1:]
            continue
        
        queue = queue[1:]
        for pair in graph[min_neigh]:
            node, weight = pair
            if node in graph:
                #node was not visited
                #insert edge at right position in queue
                insert_queue(queue, pair)            
        min_sum += min_weight        
        del graph[min_neigh]
    return min_sum
```

```
#Hackerrank main
if __name__ == "__main__":
    n, m = input().strip().split(' ')
    n, m = [int(n), int(m)]
    graph = {}
    
    for edges_i in range(m):
        
        edge = [int(edges_temp) for edges_temp in input().strip().split(' ')]
        
        if edge[0] not in graph:
            graph[edge[0]] = []
        if edge[1] not in graph:
            graph[edge[1]] = []
            
        graph[edge[0]].append((edge[1],edge[2]))
        graph[edge[1]].append((edge[0],edge[2]))
        
    start = int(input().strip())
    result = prims(n, graph, start)
    print(result)
```
