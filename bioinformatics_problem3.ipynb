# Tandem repeats

https://stepik.org/course/4377/syllabus

Tandem repeats are copies of DNA sequences which lie adjacent to each other in the same orientation (direct tandem repeats) or in the opposite direction to each other (inverted tandem repeats). Locating tandem repeats is an important task in bioinformatics, since the presence of certain tandem repeats was shown to be a marker of some wide-spread diseases. Here we consider a particular case of direct tandem repeats that occur exactly twice.

Your task is to find the longest approximate tandem repeat in given nucleotide sequences. The repeated sequences might not be exactly equal, but we require that the Levenshtein distance (https://en.wikipedia.org/wiki/Levenshtein_distance) between the parts of the repeat is not too large. We generated the tandem repeats of length at most 10000. Note that in order for us to check your solution, the length of your answer should not exceed 10000. For more details, we refer to the Scoring section below.

### <center> Input Format </center>
Each of the six tests is a FASTA file containing five nucleotide sequences. The length and properties of these sequences may vary. You can read about them in the corresponding tab. 

### <center> Output Format </center>
Your answer shoud contain five lines with three integers start, len1 and len2 each. start is the position of the first character of a tandem repeat (0-indexed),  len1 is the length of the first part and len2 is the length of the second part. All these numbers should be positive, and tandem repeat should lie entirely within the given sequence (0≤start, start+len1+len2≤L), where L is the length of the input sequence.

### <center> Sample Input </center>
Sequence 1 | Length: 11 <br>
ATACACTAGCG <br>
Sequence 2 | Length: 15 <br>
GTGGGACATACATAG <br>

### <center> Sample Output </center>
2 2 2 <br>
6 4 4 <br>

```
class Solution(object):
    @staticmethod
    def findLength(x1, x2):
        memo = [[0] * (len(x2) + 1) for _ in range(len(x1) + 1)]
        for i in range(len(x1) - 1, -1, -1):
            for j in range(len(x2) - 1, -1, -1):
                if x1[i] == x2[j]:
                    memo[i][j] = memo[i + 1][j + 1] + 1

        mx = 0
        poz = 0
        for i in range(len(x1)):
            for j in range(i, len(x2)):
                if j - i == memo[i][j]:
                    if memo[i][j] > mx:
                        mx = memo[i][j]
                        poz = i

        return poz, mx
        # return max(max(row) for row in memo)
  
    @staticmethod
    def printMatrix(a, verbose=''):
        print('--------------------')
        print(verbose)
        print('--------------------')
        for line in a:
            print(line)
        print('')
        
    @staticmethod
    def findLengthMistake(x1, x2):
        substitutions = [[0] * (len(x2) + 1) for _ in range(len(x1) + 1)]
        memo = [[0] * (len(x2) + 1) for _ in range(len(x1) + 1)]
        
        mx = 0
        poz = 0
        
        for i in range(len(x1) - 1, -1, -1):
            for j in range(len(x2) - 1, -1, -1):
                if x1[i] == x2[j]:
                    memo[i][j] = memo[i + 1][j + 1] + 1
                    substitutions[i][j] = substitutions[i+1][j+1]
                else:
                    memo[i][j] = memo[i + 1][j + 1] + 1
                    substitutions[i][j] = substitutions[i+1][j+1] + 1
        
        #Solution.printMatrix(memo,'memo')
        #Solution.printMatrix(substitutions,'substitutions')
        for i in range(0,len(x1), 1):
            for j in range(i, len(x2), 1):
                for k in range(0, len(x2) - j + 1, 1):
                    memo_aux = memo[i][j]-memo[i+k][j+k]
                    subst_aux = substitutions[i][j]-substitutions[i+k][j+k]
                    if j - i != memo_aux:
                        continue
                    if memo_aux <= mx:
                        continue    
                    if memo[i][j] > 10000:
                        continue
                    if subst_aux > 0.1*(j-i):
                        continue
                    
                    mx = memo_aux
                    poz = i

       
  
        return poz, mx
        # return max(max(row) for row in memo)
```
